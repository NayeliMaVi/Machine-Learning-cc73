```python
# Importar librerías
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset desde un archivo CSV (ajusta la ruta a tu archivo)
df = pd.read_csv('/content/proyectos_criptos_categorias.csv')

# Asegurarse de que las columnas numéricas estén en el tipo de dato correcto
df['circulating_supply'] = pd.to_numeric(df['circulating_supply'], errors='coerce')
df['total_supply'] = pd.to_numeric(df['total_supply'], errors='coerce')
df['max_supply'] = pd.to_numeric(df['max_supply'], errors='coerce')
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['market_cap'] = pd.to_numeric(df['market_cap'], errors='coerce')
df['volume_24h'] = pd.to_numeric(df['volume_24h'], errors='coerce')

# Obtener las 10 criptomonedas con mayor capitalización de mercado
top_10_market_cap = df.nlargest(10, 'market_cap')

# Obtener las 10 criptomonedas con mayor precio
top_10_price = df.nlargest(10, 'price')

# Gráfico 1: Distribución de criptomonedas por categoría
plt.figure(figsize=(10, 6))
df['category'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribución de Criptomonedas por Categoría', fontsize=14)
plt.xlabel('Categoría', fontsize=12)
plt.ylabel('Cantidad de Criptomonedas', fontsize=12)
plt.xticks(rotation=45)
plt.show()

# Gráfico 3: Nombre con Capitalización de Mercado (solo las 10 con mayor capitalización)
plt.figure(figsize=(10, 6))
plt.scatter(top_10_market_cap['name'], top_10_market_cap['market_cap'], alpha=0.5, c='orange')
plt.title('Capitalización de Mercado de Criptomonedas por Nombre (Top 10)', fontsize=14)
plt.xlabel('Nombre', fontsize=12)
plt.ylabel('Capitalización de Mercado (USD)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Gráfico 4: Nombre con Precio (solo las 10 con mayor precio)
plt.figure(figsize=(10, 6))
plt.scatter(top_10_price['name'], top_10_price['price'], alpha=0.5, c='green')
plt.title('Precio de Criptomonedas por Nombre (Top 10)', fontsize=14)
plt.xlabel('Nombre', fontsize=12)
plt.ylabel('Precio (USD)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
